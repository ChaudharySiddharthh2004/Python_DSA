```python
#  *****************************************PYTHON DSA****************************************************
# 1.Calculate String length:
def string_length(str1):
    count = 0
    for char in str1:
        count += 1
    return count

print(string_length("Siddharth"))
        
```

```python
# 2. Get first and last two characters of a string:

def string_last_two(str1):
    if len(str1) < 2: # if string ki length 2 se lam hai to blank string print kardo
        return " "

    return str1[0:2]+str1[-2:] # nhi to 0 index se 2 tak ke char aage se + aur -2 se last tak ke char print kardo

print(string_last_two("Siddharth"))
print(string_last_two("Si"))
```

```python
#3 Replace first character of string with dollar or $:

def replace_str(str1):
    char = str1[0] #extract str[0] = s into char variable

    str1 = str1.replace(char,"$") #it means replace char value that is S in the string with "$"

    return str1

replace_str("Siddharth")
```

```python
#4 Swap of 2chars in 2string:
def change_str(a,b):
    
    new_a = b[:2] + a[2:]
    new_b = a[:2] + b[2:]
    
    return new_a + " " + new_b
print(change_str("abc","xyz"))
```

```python
#5 change last char of string to "ing" or "ily":
s = input("Enter the value")

if len(s) < 3:
    print(s)
elif s.endswith("ing"):
    s = s + "ily"
    print(s)
else:
    s = s + "ing"
    print(s)
```

```python
#6 Check not or poor in the string and replace it with good:
def slice_str(s):
    not_index = s.find("not")
    poor_index = s.find("poor")
    
    
    if not_index != -1 and poor_index != -1 and poor_index > not_index:
        s = s[:not_index] + "good" + s[poor_index + len("poor"):]
    return s
    
print(slice_str("This is not that much poor "))
```

```python
#7 Remove nth character of a non-empty string:
def remove_char_string(str, n):
    first_part = str[:n]
    last_part = str[n+1:]
    
    return first_part+last_part
    
val = remove_char_string("Siddharth",1)
print(val)
```
